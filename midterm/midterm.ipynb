```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
  
a = ['SCSI 10GB', 'SCSI 20GB', 'SCSI 30GB', 'SCSI 40GB','EISA 10GB','EISA 20GB','EISA 30GB','EISA 40GB','Removable 10GB','Removable 20GB','Removable 30GB','Removable 40GB']
b_Q1 = [610000, 400000, 450000, 700000, 500000, 300000, 350000, 650000, 450000, 200000, 220000, 630000]
b_Q2 = [600000, 380000, 440000, 720500, 530000, 300200, 350000, 650000, 480000, 220000, 210000, 640000]
b_Q3 = [580000, 360000, 445000, 430000, 430000, 340000, 300000, 400000, 367000, 234000, 200000, 530000]
b_Q4 = [605000, 410000, 380000, 435000, 435000, 330000, 310000, 420000, 321000, 180000, 180000, 520000]
  
bar_width = 0.15 
  
x_Q1 =list(range(len(a)))
x_Q2 = [i+bar_width for i in x_Q1]
x_Q3 = [i+bar_width*2 for i in x_Q1]
x_Q4 = [i+bar_width*3 for i in x_Q1]

x_stick = [i+bar_width*3/2 for i in x_Q1]
  
plt.figure(figsize=(20,8),dpi=80)
plt.xticks(x_stick,a,fontsize = 10)
  
plt.bar(range(len(a)), b_Q1, width=bar_width, label='Q1',color = 'skyblue')
plt.bar(x_Q2, b_Q2, width=bar_width, label='Q2', color = 'salmon')
plt.bar(x_Q3, b_Q3, width=bar_width, label='Q3', color = 'orange')
plt.bar(x_Q4, b_Q4, width=bar_width, label='Q4',color = 'silver')
  
plt.legend(fontsize = 10)
plt.title("Price per product per quarter",fontsize = 15)
plt.xlabel("Product")
plt.ylabel("Dollar")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5), dpi=100)
game = ['Q1','Q2','Q3','Q4']
D = [-15000, -13000, -10000, -2000]
F = [18000, 37000,-37000,40000]
H = [-4000, -8000, -10000, 17000]
I = [-25000, 40000, 101000, 30000]
plt.plot(game, D, c='red', label="Distribution", linewidth = 2)
plt.plot(game, F, c='green', linestyle='--', label="Facilities",linewidth = 2)
plt.plot(game, H, c='blue', linestyle='-.', label="Human Resources",linewidth = 2)
plt.plot(game, I, c='peru', linestyle=':', label="Information Systems",linewidth = 2)
plt.scatter(game,D , c='salmon',s = 20,marker='o')
plt.scatter(game, F, c='green',s = 20,marker='<')
plt.scatter(game, H, c='powderblue',s=20,marker="v")
plt.scatter(game, I, c='gold',s = 20, marker = 's')
plt.legend(loc='best')
plt.yticks(fontsize = 10)
plt.xticks(fontsize = 10)
plt.grid(True, linestyle='--', alpha=1)
plt.xlabel("Department", fontdict={'size': 10})
plt.ylabel("Deviation", fontdict={'size': 10})
plt.title("The Budget Deviation of Different Department", fontdict={'size': 10})
plt.show()
```
